```python
import cv2
from google.colab.patches import cv2_imshow

image_path = "sample5.jpg" # Corrected filename from "sample 5.jpg" to "sample5.jpg"
img = cv2.imread(image_path)

if img is None:
    print(f"Error: The image file '{image_path}' could not be loaded.")
    print("Please ensure the image file is present and the path is correct.")
else:
    big = cv2.resize(img, None, fx=1.5, fy=1.5)
    small = cv2.resize(img, None, fx=0.5, fy=0.5)

    # Using cv2_imshow for displaying images in Google Colab
    cv2_imshow(img)
    cv2_imshow(big)
    cv2_imshow(small)

    # cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow in Colab.
```
