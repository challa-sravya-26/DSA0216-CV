```python
# ==========================================
# Boundary Detection using Convolution Kernel
# Google Colab - Single Cell
# ==========================================

import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

# Step 1: Upload Image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Step 2: Read Image and Convert to Grayscale
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Define Convolution Kernel (Laplacian)
kernel = np.array([[0, -1,  0],
                   [-1, 4, -1],
                   [0, -1,  0]])

# Step 4: Apply Convolution
boundary = cv2.filter2D(gray, -1, kernel)

# Step 5: Display Results
print("Original Grayscale Image")
cv2_imshow(gray)

print("Boundary Image")
cv2_imshow(boundary)
```
