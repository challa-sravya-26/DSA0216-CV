```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 24 - Rotation")

img = cv2.imread("sample 10.png")

if img is None:
    print("Upload sample 10.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    h,w = img.shape[:2]
    M = cv2.getRotationMatrix2D((w/2,h/2),45,1)
    rotated = cv2.warpAffine(img,M,(w,h))

    print("OUTPUT IMAGE:")
    cv2_imshow(rotated)
```
