```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

print("Experiment 19 - Erosion")

img = cv2.imread("sample 7.png")

if img is None:
    print("Upload sample 7.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    kernel = np.ones((5,5), np.uint8)
    erosion = cv2.erode(img, kernel, iterations=1)

    print("OUTPUT IMAGE:")
    cv2_imshow(erosion)
```
