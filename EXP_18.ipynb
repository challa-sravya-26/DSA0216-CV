```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

print("Experiment 18 - Dilation")

img = cv2.imread("sample 8.jpg")

if img is None:
    print("Upload sample 8.jpg")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    kernel = np.ones((5,5), np.uint8)
    dilated = cv2.dilate(img, kernel, iterations=1)

    print("OUTPUT IMAGE:")
    cv2_imshow(dilated)
```
