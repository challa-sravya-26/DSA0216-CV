```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image_path = "sample 6.png" # Corrected filename from .jpg to .png
img = cv2.imread(image_path)

if img is None:
    print(f"Error: The image file '{image_path}' could not be loaded.")
    print("Please ensure the image file is present and the path is correct.")
else:
    (h, w) = img.shape[:2]

    M = np.float32([[1, 0, 100], [0, 1, 50]])

    shifted = cv2.warpAffine(img, M, (w, h))

    cv2_imshow(img)
    cv2_imshow(shifted)

    # cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow in Colab.
```
