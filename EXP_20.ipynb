```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

print("Experiment 20 - Opening")

img = cv2.imread("sample 10.png")

if img is None:
    print("Upload sample 10.png")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    kernel = np.ones((5,5), np.uint8)
    opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

    print("OUTPUT IMAGE:")
    cv2_imshow(opening)
```
