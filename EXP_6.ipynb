```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Step 1: Create Video
height, width = 400, 600
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('generated_video.mp4', fourcc, 20.0, (width, height))

for i in range(100):
    frame = np.zeros((height, width, 3), dtype=np.uint8)

    # Draw moving circle
    cv2.circle(frame, (50 + i*5, 200), 40, (0,255,0), -1)

    out.write(frame)

out.release()

print("Video Created Successfully!")

# Step 2: Play Video (Slow Motion)
cap = cv2.VideoCapture('generated_video.mp4')

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2_imshow(frame)
    cv2.waitKey(200)   # increase delay for slow motion

cap.release()
```
