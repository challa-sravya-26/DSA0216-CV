```python
import cv2
from google.colab.patches import cv2_imshow

print("Experiment 26 - Image Flipping")

img = cv2.imread("sample 6.jpg")

if img is None:
    print("Upload sample 6.jpg")
else:
    print("INPUT IMAGE:")
    cv2_imshow(img)

    flipped = cv2.flip(img, 1)

    print("OUTPUT IMAGE:")
    cv2_imshow(flipped)
```
