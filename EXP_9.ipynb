```python
import cv2
from google.colab.patches import cv2_imshow # Import cv2_imshow

img = cv2.imread("sample 9.png") # Changed from sample 9.jpg to sample 9.png

if img is None:
    print("Error: Could not load image 'sample 9.png'. Please ensure the file exists and the path is correct.")
else:
    (h, w) = img.shape[:2]

    center = (w//2, h//2)

    clock = cv2.getRotationMatrix2D(center, -45, 1.0)
    anticlock = cv2.getRotationMatrix2D(center, 45, 1.0)

    rot1 = cv2.warpAffine(img, clock, (w, h))
    rot2 = cv2.warpAffine(img, anticlock, (w, h))

    cv2_imshow(rot1) # Changed cv2.imshow to cv2_imshow
    cv2_imshow(rot2) # Changed cv2.imshow to cv2_imshow

    # cv2.waitKey(0) # Not needed with cv2_imshow
    # cv2.destroyAllWindows() # Not needed with cv2_imshow
```
