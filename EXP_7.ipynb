```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Simulate live frames
for i in range(50):
    frame = np.zeros((400,600,3), dtype=np.uint8)

    cv2.putText(frame, "Simulated Webcam Frame",
                (100,200),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (255,255,255),
                2)

    cv2.rectangle(frame, (50+i*5, 100), (150+i*5, 200), (0,255,0), -1)

    cv2_imshow(frame)
    cv2.waitKey(100)
```
